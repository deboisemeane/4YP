```python
#Load Hypnogram
def load_hypnogram(patient, xunits="e"):
  #First number: patient in subgroup3, Second number: human expert who created the hypnogram
  y =  np.loadtxt(hypnograms_1[patient-1], dtype=int)
  #Choose time units "e" for epochs, "s" for seconds, "h" for hours
  xunits = "e"
  if xunits=="h":
    x = np.arange(0,len(y))*30/(60*60)#Time in hrs
  elif xunits=="e":
    x = np.arange(0,len(y))#30s Epochs
  elif xunits=="s":
    x = np.arange(0,len(y))*30 #Time in s

  #Transform stage data to be lower number = deeper sleep
  for i in range(len(y)):
      if y[i] == 5:  # 5 = REM
          y[i] = 3    #Now, 3 = REM
      elif y[i] == 3:# 3 = N3
          y[i] = 0    #Now, 0 = N3
      elif y[i] == 2:# 2 = N2
          y[i] = 1    #Now, 1 = N2
      elif y[i] == 1:# 1 = N1
          y[i] = 2    #Now, 2 = N1
      elif y[i] == 0:# 0 = W
          y[i] = 4    #Now, 4 = W

  stage = y
  time = x
  return time, stage

#time, stage = load_hypnogram(patient)
```

```python
#Ho-Sheng Data Preprocessing Pipeline

#EPOCH LENGTH MUST BE DIVISIBLE BY windowlength*(1-overlap)
def overlay_windows(patient, window_length=10, overlap=0.5, channel="C4-A1", epoch_length=30, eeg=eeg):
  raw_data = mne.io.read_raw_edf(eeg[patient-1])  #Load raw data
  print(raw_data.info["ch_names"])
  channel_data = raw_data.pick_channels([channel])
  #Overlay windows
  detrend=0 #0 for dc detrend

  #Split each 30s epoch into five 10s windows overlaying by 50%, to preserve attenuated features at edge of hamming windows.
  t, labels = load_hypnogram(patient)
  metadata = pd.DataFrame({"Sleep Stage" : labels})
  events_30s = mne.make_fixed_length_events(channel_data, duration=epoch_length)
  epochs_30s = mne.Epochs(channel_data, events_30s, tmin=0, tmax=epoch_length, metadata=metadata, preload=False, detrend=detrend, baseline=None)


  #Create labels for the overlapping windows, based on the stage they start in??
  window_labels = np.repeat(labels, int(epoch_length / (window_length*(1-overlap))) )
  metadata = pd.DataFrame({"Sleep Stage" : window_labels})
  events_10s = mne.make_fixed_length_events(channel_data, duration=window_length*(1-overlap))
  epochs_10s = mne.Epochs(channel_data, events_10s, tmin=0, tmax=window_length, preload=True, metadata=metadata, baseline=None)
  #print("Windows overlapping correctly: "+str(epochs_10s.get_data()[1][0][0] == epochs_10s.get_data()[0][0][int(_/2)])) #checks overlapping corectly
  return epochs_30s, epochs_10s


```

```python
#Downsample and LPF
############
#Downsample to improve computational efficiency
def downsample(epochs, downsample_factor=5):
  sfreq = epochs.info["sfreq"]
  ds_epochs_10s = epochs.copy()
  return ds_epochs_10s.resample(sfreq/downsample_factor) #has built in LPF to avoid aliasing artifacts

###########
#LPF - not sure this really does anything after downsampling, since the highest frequency we can now represent is only 20Hz
def lpf(epochs, cutoff=19):
  #stop_freq = 20 #stopband i.e. transition frequency of 5

  #Define the iir params
  iir_params = {
      'ftype': 'butter',
      'order': 2,  # 2th order filter (passed twice gives 80db/decade stopband)
      'output': 'sos'}  # Second-order sections representation, more numerically stable

  method='iir' #Forward-backward filtering using filtfilt()
  phase='zero' #Forward backward filtering?? Not sure difference between 'zero-double'
  lpf_epochs = epochs.copy()
  lpf_epochs.filter(l_freq=None, h_freq=cutoff, method=method, iir_params=iir_params, phase=phase )

  return lpf_epochs

#ds_epochs_10s = downsample(epochs_10s)
#lpf_epochs_10s = lpf(ds_epochs_10s)
```

```python
#Hamming Window and FFT linearised between 0 and 1
#
###########
def apply_hamming_fft(epochs):
  windowsize = epochs.get_data().shape[2] #no. of points in each epoch
  hamming_window = np.hamming(windowsize)

  windowed_data = epochs.get_data() * hamming_window
  fft_data = np.abs(np.fft.rfft(windowed_data, axis=2))                          #Take absolute values
  linearised_fft_data = fft_data / np.max(fft_data, axis=2, keepdims=True)       #Divide by max value in each epoch

  sfreq=epochs.info['sfreq']
  nyquist_freq = sfreq / 2  #maximum frequency in the fft
  freqs = np.linspace(0, nyquist_freq, len(fft_data[0][0])) #Calculate frequencies (evenly spaced from 0 to nyquist frequency)

  return freqs, linearised_fft_data


#Select Frequency features
#1 feature per 1Hz interval
###############
def select_freq_features(fft_data, freqs, bin_freq=1):
  maxfreq = freqs[-1]
  f_interval = freqs[1]-freqs[0]
  bin_size = int(bin_freq / f_interval)

  n_epochs, n_channels, n_freqs = fft_data.shape

  n_bins = int(n_freqs / bin_size)  #REQUIRES BIN SIZE TO PERFECTLY DIVIDE INTO FREQS

  binned_data = np.zeros((int(n_epochs), int(n_channels), int(n_bins)))
  feature_freqs = np.zeros(n_bins)
  for i in range(0, n_bins):
    start_idx = i*bin_size
    end_idx = start_idx + bin_size
    binned_data[:,:,i] = np.mean(fft_data[:,:,start_idx:end_idx], axis=2)
    feature_freqs[i] = np.mean(freqs[start_idx:end_idx])
  return feature_freqs, binned_data
```

```python
#Comparing delta theta alpha beta power by sleep stage
def get_power_fractions(patient,):
  #Load data
  epochs_30s = mne.read_epochs('/content/gdrive/MyDrive/Colab Notebooks/4YP/Datasets/ISRUC/SubgroupIII/Saved Epochs/{patient}_30s-epo.fif'.format(patient=patient))
  epochs_10s = mne.read_epochs('/content/gdrive/MyDrive/Colab Notebooks/4YP/Datasets/ISRUC/SubgroupIII/Saved Epochs/{patient}_10s-epo.fif'.format(patient=patient))
  epochs_10s.load_data()
  ds_epochs_10s = downsample(epochs_10s)
  lpf_epochs_10s = lpf(ds_epochs_10s)
  freqs, linearised_fft_data = apply_hamming_fft(lpf_epochs_10s)
  labels = lpf_epochs_10s.metadata["Sleep Stage"]
  #Epoch indices for corresponding stages N3 N2 N1 REM W
  stage_idx = np.zeros((5, len(labels)))
  for i in range(5):
    stage_idx[i] = labels == i
  #Frequency range indices for corresponding clinical bands
  delta_idx = np.logical_and(np.greater_equal(freqs,0.5), np.less(freqs,4))
  theta_idx = np.logical_and(np.greater_equal(freqs,4), np.less(freqs,8))
  alpha_idx = np.logical_and(np.greater_equal(freqs,8), np.less(freqs,12))
  beta_idx = np.logical_and(np.greater_equal(freqs,12), np.less(freqs,30))
  #Power fractions for each window
    #Delta, theta, alpha, beta
  delta_fft = linearised_fft_data * delta_idx #Shape (n_windows, n_channels, n_freqs), zeros outside the desired frequency range
  theta_fft = linearised_fft_data * theta_idx
  alpha_fft = linearised_fft_data * alpha_idx
  beta_fft = linearised_fft_data * beta_idx

  fft = np.squeeze(np.array([delta_fft, theta_fft, alpha_fft, beta_fft])) #Shape (n_bands, n_windows, n_freq)

  powers = np.sum(fft*fft, axis=2)       #Total power for each window, shape (n_bands, n_windows)

  power_fractions = powers / np.sum(powers, axis=0)  #Power fractions for each window, shape (n_bands, n_windows)

  #Powers for each sleep stage, for each window
  stage_powers = np.flip(np.expand_dims(powers, 1) * np.expand_dims(stage_idx, 0), axis=1) #Shape (n_bands (4), n_stages (5), n_windows)
                                                                                          #axis0: bands delta->beta
                                                                                          #axis1: stages 0->4 | W->N3

  #Average powers for each sleep stage
  avg_stage_powers = np.average(stage_powers, axis=2)   #Averaged over windows corresponding to each sleep stage. Shape (n_bands, n_stages)
  avg_stage_power_fractions = avg_stage_powers / np.sum(avg_stage_powers, axis=0)
  #Average Power fractions for each sleep stage, across all windows
  #avg_power_fractions = np.average(stage_power_fractions, axis=2) #Shape (n_bands(4), n_stages(5))
  print(avg_stage_powers.shape)
  return avg_stage_power_fractions, power_fractions
```
